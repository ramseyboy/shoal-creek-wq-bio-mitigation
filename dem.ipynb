```python
from matplotlib import pyplot
from rasterio.plot import show, show_hist
import rasterio.mask

with rasterio.open("data/shoal_creek_dem_10m.tif") as dem:
    fig, (axrgb, axhist) = pyplot.subplots(1, 2, figsize=(33,33))
    show(dem, ax=axrgb)
    show_hist(dem, bins=50, histtype='stepfilled', lw=0.0, stacked=False, alpha=0.3, ax=axhist)
    pyplot.show()
```
